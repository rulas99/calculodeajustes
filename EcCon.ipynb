```python
import pandas as pd
import numpy as np
```

```python
# Definimos una función que genera un dataframe donde cada columna correspondera al segmento.
# n es la cantidad de puntos

def cols(n):  
    pts=np.arange(1,n+1)
    cols=[]

    for i in range(len(pts)):

        if i!=len(pts)-1:

            x=pts[i]

            for j in pts[i+1:]:

                seg=str(x)+'-'+str(j)

                cols.append(seg)
                
    return pd.DataFrame(columns=cols)
```

```python
# Definimos la funcón que va a reellanar mi tabla de ecuaciones
# n es la cantidad de puntos

def matr(n):
    sn=str(n)
    mat=cols(n)
    inis=[i for i in np.array(mat.columns)[:-1] if (i.split('-')[1]!=sn) and (i.split('-')[0]=='1')]
    for i in inis:
        mat=mat.append(filec(i,n))

    return mat.reset_index(drop=True)
```

## Algorítmo que genera las ecuaciones de la forma vista en clase

```python
# Definimos la función que se va a encargar de crear las ecuaciones de condición
# n es la cantidad de puntos

def filec(ini,n):

    df=cols(n)
    df=df.append(pd.DataFrame([np.repeat(1,len(df.columns))],columns=df.columns))

    ecs=cols(n)
    sta=ini.split('-')[0]
    fin=sta+'-'+str(n)
    col=[ini,fin]
    cof=[df[ini][0],-df[fin][0]]

    sta=ini
    c=0

    for i in range(int(ini.split('-')[1]),n):

        ini=sta
        col=col[:2]
        cof=cof[:2]

        while int(ini.split('-')[1])!=n:

            ix=ini.split('-')
            xi=int(ix[0])
            xf=int(ix[1])
            ini=str(xf)+'-'+str(xf+1+c)


            if ((xf+1+c)>=n):
                ini=str(xf)+'-'+str(n)

            cof.append(df[ini][0])
            col.append(ini) 

        ecs=ecs.append(pd.DataFrame(data=[cof],columns=col)) 
        c+=1

    return ecs.fillna(0)
```

## Resolución para una base de calibración SIN pesos

$V=B^T(B^TB)^{-1}W$ 

Donde $W=C-BL$

```python
seg=pd.read_csv('/Users/Raul/Desktop/Tarea12.csv')
seg
```

```python
B=np.array(matr(12), dtype='float')
L=np.array(seg.distancia, dtype='float')
W=np.repeat(0,len(B))-np.dot(B,L)

sol=np.dot(np.dot(B.transpose(),np.linalg.inv(np.dot(B,B.transpose()))),W)
Laj=np.round(L+sol,3)
```

```python
seg['Ajuste']=np.round(sol,3)
seg['DistAj']=Laj
```

## Resolución para una base de calibración CON pesos

$V=P^{-1}B^T(BP^{-1}B^T)^{-1}W$ 

Donde $W=C-BL$

```python
med=pd.read_csv('pes.csv',index_col='segmento')
med
```

```python
# Matriz de pesos, haciendo pruebas me percate que da igual si el numerador es 1 o un 1000000

P=np.diag([1/np.std(med.loc[i],ddof=1) for i in med.index])
P
```

```python
B=np.array(matr(6), dtype='float')
L=np.array([np.mean(med.loc[i]) for i in med.index], dtype='float')
W=-np.dot(B,L)
PI=np.linalg.inv(P)
sol=np.dot(PI,np.dot(np.dot(B.transpose(),np.linalg.inv(np.dot(np.dot(B,PI),B.transpose()))),W))
Laj=np.round(L+sol,3)
```

```python
med['$\mu$']=[np.round(np.mean(med.loc[i]),3) for i in med.index]
med['$\epsilon$']=[np.round(np.std(med.loc[i],ddof=1),3) for i in med.index]
med['$Var$']=np.round(sol,3)
med['$Ajuste$']=Laj
```
